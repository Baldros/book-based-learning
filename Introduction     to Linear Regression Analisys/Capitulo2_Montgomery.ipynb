Capitulo 2:

    O capitulo 2 do livro do Montgomery trata da Regressão Linear Normal
    simples. A Regressão Linear Respeita alguns pressupostos importantes.

Modelo Teórico:

$Y_{i \in \mathbb{N}^*} = \beta_0+\beta_1x_{i \in \mathbb{N}^*}+\epsilon_{i \in \mathbb{N}^*}$

Onde:

* $\epsilon_{i \in \mathbb{N}^*} \sim N(0,\sigma^2)$:

    Se os erros não tiverem distribuição normal, já impossibilita a utilização
    dessa técnica para o aprendizado. Isso é algo que muitos não falam, mas é
    um pressuposto que precisa ser respeitado para um modelo preciso.

* $Y_{i \in \mathbb{N}^*}$:

    Não precisam ser identicamente distribuidos mais precisam ser independentes
    entre SI, sendo dependentes apenas da regressora. A aleatoriedade gerada
    pelo erro é transportada para as saídas.

* $x_{i \in \mathbb{N}^*}$:

    São os dados da nossa variável regressora. Nos modelos de aprendizado
    estatístico, precisamos dos dados para ajustar as nossas funções de
    predição.

*$\beta_0 e \beta_1$:

    São os coeficientes da nossa regressão. Para o caso simples, eles possuem,
    um nome bem específico, são, respectivamente interceptor e coeficiente
    angular. Por vezes, o coeficiente angular da reta de regressão também é
    chamado de correlação empírica, verémos o porquê mais para a frente.

# Questão 1:
A Tabela B.1 fornece dados sobre o desempenho das 26 equipes da National Football League em 1976. Suspeita-se que o número de jardas ganhas em corridas pelos adversários (x8) tenha um efeito sobre o número de jogos ganhos por uma equipe (y).

```r
# Analisando Visualmente os dados:
plot(x8, y, pch = 20, cex = 2, xlab = 'Jardas ganhas em corridas pelos adversários (temporada)', ylab = 'Partidas vencidas')
```

    Vemos um certo nivel de associação linear, mas negativa, ou seja,
    porém com uma correlação (empírica) negatíva.

**Avaliando a Correlação**

Correlação de Pearson:
$r=\frac{\sum_{i=1}^ny_i(x_i-\bar{x})}{\left[\sum_{i=1}^n(x_i-\bar{x})^2 \sum_{i=1}^n(Y_i-\bar{y})^2\right]^{\frac{1}{2}}} = \frac{S_{xy}}{[S_{xx}SS_T]^{\frac{1}{2}}}$

```r
# Correlação de Pearson:
cor(x8,y)
```

a. Ajuste um modelo de regressão linear simples relacionando o número de jogos ganhos (y) com as jardas ganhas em corridas pelos adversários (x8).

```r
# Construção do Modelo de Regressão:
modelo <- lm(y ~ x8);modelo
```

```r
# Informações estatísticas do Modelo:
summary(modelo)
```

**Interpretação:**

* Intercepto (coeficiente $\beta_0$):


    O intercepto é estimado em aproximadamente 21.79. Isso significa que,
    quando o número de jardas ganhas em corridas pelos adversários é zero,
    espera-se que a equipe ganhe cerca de 21.79 jogos em uma temporada de
    14 jogos.

* Coeficiente de x8 (coeficiente $\beta_1$):


    O coeficiente associado a x8 é estimado em -0.007025. Isso indica que,
    em média, para cada jarda adicional ganha em corridas pelos adversários,
    espera-se que a equipe ganhe aproximadamente 0.007025 jogos a menos em
    uma temporada de 14 jogos.

* P-valor:
    

    O p-valor associado ao coeficiente de x8 é muito baixo (7.38e-06), o
    que sugere que há uma forte evidência estatística de que o coeficiente
    é significativamente diferente de zero.

* R-quadrado:


    O R-quadrado (coeficiente de determinação) do modelo é aproximadamente
    0. 545, o que significa que cerca de 54.5% da variabilidade no número
    de jogos ganhos é explicada pela variabilidade nas jardas ganhas em
    corridas pelos adversários.

* F-statistic:


    O valor F do teste de significância global do modelo é 31.1, com
    um p-valor muito baixo (7.381e-06), indicando que o modelo como
    um todo é estatisticamente significativo.
